```python
#importar bibliotecas
import pyautogui
import pyperclip
import time


# Adicionar pausa entre os codigos para a automaçao fluir melhor
pyautogui.PAUSE = 1

# Passo 1 - Entrar no sistema da empresa
#Abrir nova aba no chrome
pyautogui.hotkey("ctrl", "t")

#Usar o pyperclip para copiar o link e colar na nova aba
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

#Adicionar uma pausa exclusiva para essa linha de codigo, para aguardar carregar o site
time.sleep(5)


# Passo 2 - Navegar no Sistema ate a base de vendas
#entrar no menu explorar
pyautogui.click(x=341, y=301, clicks=2)
time.sleep(2)

# Passo 3 - Fazer o download da base de vendas
#clicar no arquivo
pyautogui.click(x=375, y=372)
time.sleep(2)

#clicar nos tres pontos para download
pyautogui.click(x=1159, y=192)
time.sleep(2)

#clicar em fazer download
pyautogui.click(x=989, y=565)
time.sleep(5)
```

```python
# Passo 4 - Importar a base para o Python
import pandas

tabela = pandas.read_excel(r"C:\Users\Ricardo\Downloads\Vendas - Dez.xlsx")
display(tabela)

```

```python
# Passo 5 - Calcular os indicadores da empresa
Faturamento = tabela["Valor Final"].sum()
print(Faturamento)
Quantidade = tabela["Quantidade"].sum()
print(Quantidade)
```

```python
# Passo 6 - Enviar e-mail para a diretoria com os indicadores de vendas
#abrir nova aba
pyautogui.hotkey("ctrl", "t")

#colar link do email
pyperclip.copy("https://mail.google.com/mail/u/6/#inbox")
#colar
pyautogui.hotkey("ctrl", "v")
#entrar
pyautogui.press("enter")
#pausa
time.sleep(5)

# clicar em escrever
pyautogui.click(x=77, y=200)
time.sleep(3)

# Preencher informações do e-mail
pyautogui.write("contatodosordili@gmail.com") #escrever email
pyautogui.press("tab") #selecionar email
pyautogui.press("tab") #pular para assunto
pyperclip.copy("Relatório de Vendas") #escrever assunto
pyautogui.hotkey("ctrl", "v") #colar assunto
pyautogui.press("tab") #pular para corpo do email

#escrever o email na variavel texto

texto = f"""Prezados, Bom dia.

Segue abaixo relatório diário de vendas.
Quantidade vendida: {Quantidade:,}
Faturamento Total: R$ {Faturamento:,.2f}

Qualquer dúvida estou a disposição;
Att."""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

time.sleep(2)

pyautogui.hotkey("ctrl", "enter")
```
